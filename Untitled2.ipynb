```python
df = pd.read_csv('/content/batter.csv')
```

```python
df.head(100).sample(25, random_state=0)
```

```python
sample = df.head(100).sample(25, random_state=0)
```

```python
plt.figure(figsize= (10,10))

plt.scatter( sample['avg'], sample['strike_rate'],s = sample['runs'], marker= '*')

plt.axhline(130, color = 'yellow')

for i in range (sample.shape[0]):
  plt.text(sample['avg'].values[i],sample['strike_rate'].values[i],sample['batter'].values[i])

plt.show()
```

 plt.text(x, y, label) requires you to tell it exactly what:

x position (→ sample['avg'].values[i])

y position (↑ sample['strike_rate'].values[i])

and the label/text (→ sample['batter'].values[i]) you want to display.

```python
fig, ax = plt.subplots(nrows= 2, ncols=1, sharex= True, figsize=(10,6))

ax[0].scatter(df['avg'], df['strike_rate'],color = 'red')
ax[1].scatter(df['avg'], df['runs'])

ax[0].set_title('avg vs rate')
ax[0].set_ylabel('rate')


ax[1].set_title('avg vs runs')
ax[1].set_ylabel('runs')
ax[1].set_xlabel('avg')
```

📋 List of Common plot() Parameters (Pandas)
Parameter	Description
kind	Type of plot ('line', 'bar', 'hist', 'box', 'scatter', etc.)
x, y	Column names for axes
color	Color of plot elements
marker	Marker symbol (for scatter or line plots)
s	Size of markers (scatter only)
figsize	Tuple for plot size, e.g., (10, 6)
title	Plot title
xlabel, ylabel	Labels for axes
grid	Show grid (True/False)
legend	Show legend
alpha	Transparency (0 to 1)
linewidth / lw	Width of lines
linestyle / ls	Line style: '-', '--', ':', etc.
cmap	Colormap for color gradients
